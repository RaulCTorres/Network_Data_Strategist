# Network Data Strategist

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import from csv
attendance_df = pd.read_csv('Attendance.csv')
#Original df
attendance_df
```

```python
attendance_df['School ID'] = attendance_df['School ID'].fillna(method='ffill')
attendance_df['Network'] = attendance_df['Network'].fillna(method='ffill')
```

```python
#Drop rows with any null values in '2015', '2016', '2017'
attendance_df = attendance_df.dropna(subset=['2015', '2016', '2017'])
```

```python
network_12_attendance_df = attendance_df[attendance_df['Network'] == 'Network 12']
network_12_attendance_df
```

```python
# Calculate the mean attendance for each year
average_attendance_per_year = network_12_attendance_df[['2015', '2016', '2017']].mean()
```

```python
# Plot the average attendance trend
plt.figure(figsize=(10, 6))
plt.plot(average_attendance_per_year.index, average_attendance_per_year.values, marker='o', linestyle='-', color='b')
plt.title('Average Attendance Trend in Network 12 (2015-2017)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Attendance Rate (%)', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
