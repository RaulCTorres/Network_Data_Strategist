```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import from csv
nwda_growth_df = pd.read_csv('NWEA_Growth.csv')
#original df
nwda_growth_df.head()
```

```python
#drop null values
nwda_growth_df.dropna(inplace=True)
```

```python
#create dataframe for only combined grades
combined_df = nwda_growth_df[nwda_growth_df['Grade'].str.lower().str.contains('combined')]
combined_df.head()
```

```python
#dataframe without combined grades
clean_df = nwda_growth_df[~nwda_growth_df['Grade'].str.lower().str.contains('combined')]
clean_df.head()
```

```python
combined_df.columns.to_list()
```

```python
plt.scatter(x= '2014 # Tested', y= '2014 % Students Making National Average Growth',  data=combined_df)
```

```python
sns.regplot(x='2014 # Tested', y='2014 % Students Making National Average Growth', data=combined_df, scatter_kws={'s': 100}, line_kws={'color': 'red'})
```
